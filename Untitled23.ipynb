```python
from gtts import gTTS
from IPython.display import Audio

def speak(text):
    tts = gTTS(text=text, lang='en')
    tts.save("output.mp3")
    return Audio("output.mp3", autoplay=True)

# Example usage
speak("Hello, how can I assist you today by?")
```
